## Transformando listas em Pandas Series

```python
import pandas as pd
idades_lista = [25, 30, 22, 28]
idades_series = pd.Series([25, 30, 22, 28])
print(f"Lista:\n{idades_lista}\n")
print(f"Series:\n{idades_series}")
```
